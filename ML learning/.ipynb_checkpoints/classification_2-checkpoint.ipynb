```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X,y = mnist.data,mnist.target
```

```python
import matplotlib.pyplot as plt

def plot_digit(image_data):
    image = image_data.reshape(28,28)
    plt.imshow(image,cmap = "binary")
    plt.axis("off")

some_digit = X[0]
plot_digit(some_digit)
plt.show()
```

```python
y[0]
```

```python
X_train,X_test,y_train,y_test = X[:60000],X[60000:],y[:60000],y[60000:]# already suffled -- gauratees cross validation fold
```

```python
for i in range(13):
    plot_digit(X[i])
    plt.show()
```

## ## Traning a Binary Classifier for 5-detector

```python
(y_train == '5') 
```

```python
y_train_5 = (y_train == '5')
y_test_5 = (y_test == '5')
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state = 42)
sgd_clf.fit(X_train,y_train_5)
```

```python
sgd_clf.predict([some_digit])
```

## Performance Measure 
### Measure of Accurarcy using Cross-Validation 

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(sgd_clf,X_train,y_train_5,cv = 3,scoring ="accuracy")
```

```python
from sklearn.dummy import DummyClassifier

dummy_clf = DummyClassifier()
dummy_clf.fit(X_train,y_train_5)
print(any(dummy_clf.predict(X_train)))
```

```python
cross_val_score(dummy_clf,X_train,y_train_5,cv = 3,scoring= "accuracy")
```

This because only 10% of the dataset is 5 and guessing that image is not 5 is right about 90%
